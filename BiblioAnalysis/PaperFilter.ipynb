```python
import pandas as pd

# Load the scopus.csv file
data = pd.read_csv('biblio/scopus1837.csv')

# Display the first few rows to understand the structure of the data
print(data.head())

# Define a function to filter rows related to education
def is_relevant_to_education(title):
    keywords = ['education', 'teaching', 'teacher', 'classroom', 'students', 'curriculum', 'pedagogy','campus', 'academic','e-learning','wrote']
    text = f"{title}".lower()
    return any(keyword in text for keyword in keywords)

# Apply the filter to the dataset
if 'Title' in data.columns and 'Abstract' in data.columns:
    filtered_data = data[data.apply(lambda row: is_relevant_to_education(row['Title']), axis=1)]
else:
    raise ValueError("The required columns 'Title' and 'Abstract' are not present in the dataset.")

# sort and save the filtered data to a new CSV file
sorted_data = filtered_data.sort_values(by='Cited by', ascending=False)
sorted_data.to_csv('res/filtered_scopus.csv', index=False)

print("Filtered data has been saved to 'filtered_scopus.csv'")

```

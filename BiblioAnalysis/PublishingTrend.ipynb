```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('lot_key_all.csv')

# Ensure the column containing the year is correctly named and available
if 'Year' not in df.columns:
    print("Column 'Year' not found in the file.")
else:
    # Count the number of articles published each year
    year_counts = df['Year'].value_counts().sort_index()
    
    # filte the range of year
    year_counts = year_counts[(year_counts.index >= 2010) & (year_counts.index <= 2024)]

    # Plot the data
    plt.figure(figsize=(10, 6))
    plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-')
    plt.title('Number of Articles Published Per Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Articles')
    
    # limit the x-axis
    plt.xticks(range(2010,2025))
    
    plt.grid(True)
    plt.show()
```
